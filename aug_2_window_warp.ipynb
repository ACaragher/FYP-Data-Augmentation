# Augmentation Method #2 - Window Warping
<br><br>
- Implementation of second augmentation method, Window Warping
- Take a panas DataFrame and stretch the time series in some window, returning the resulting time series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from random import randrange
import rocket
import mpdatasets
```

```python
train_x, train_y, test_x, test_y = mpdatasets.load_datasets()
```

<br><br>
- A window some percentage of the length of the time series is selected
- The window is centred in the middle of the time seres
- Within this window, the time series is stretched by some multiplier (default x2)

```python
def window_warp(dfx, window_size, multiplier, stepped=False):
    warped_series = []

    ts_length = len(dfx.iloc[0][0])

    for i in range(dfx.shape[0]):

        window_start = ts_length // 2 - window_size // 2

        window_end = window_start + window_size
        body_parts = []
        
        
        for j in range(dfx.shape[1]):
            
            # Add any values form 0 to beginning of the window to the aug_time_series
            # Then, for each value in the window, add it'multiplier' times to aug_time_series
            # Finally, add any values after the window till the end to the aug_time_series
            aug_time_series = dfx.iloc[i][j][:window_start] 
            if stepped:
                for k in range(window_start, window_end):
                    for m in range(0,multiplier):
                        aug_time_series = pd.concat([aug_time_series, dfx.iloc[i][j][k:k+1]], axis=0, ignore_index=True)
                        
                aug_time_series = pd.concat([aug_time_series, dfx.iloc[i][j][window_end:]], axis=0, ignore_index=True)
        
                body_parts.append(aug_time_series)

            # To warp: take two adjacent values. Add 'multiplier' number of new values 
            # between them. The values are linear between the two original values
            else:
                for k in range(window_start, window_end):
                    #
                    difference = dfx.iloc[i][j][k+1].item() - dfx.iloc[i][j][k].item()
                    change_amount = 0
                    # The first value is added to the time series on the first iteration
                    # On subsequent iterations, difference/multiplier is added each time to the
                    # previously added value
                    for m in range(0,multiplier):
                        new_value = pd.Series([dfx.iloc[i][j][k].item() + change_amount])
                        aug_time_series = pd.concat([aug_time_series, new_value], axis=0, ignore_index=True)
                        change_amount += difference/multiplier

                aug_time_series = pd.concat([aug_time_series, dfx.iloc[i][j][window_end:]], axis=0, ignore_index=True)

                body_parts.append(aug_time_series)

        warped_series.append(body_parts)
        
    warped_frame = pd.DataFrame(warped_series)
    warped_frame.columns = dfx.columns
    return warped_frame
```

<br><br>
- The test and training datasets' time series must also be the same length as the augmented set, therefore their time series are also padded
- Series are padded with their first value at the front and final value at the end evenly to make it the correct length

```python
def pad_series(dfx, pad_size, multiplier):
    padded_series_list = []

    for i in range(dfx.shape[0]):

        padded_series = []

        for j in range(dfx.shape[1]):
            
            first_value = dfx.iloc[i][j][0].item()
            final_value = dfx.iloc[i][j][-1:].item()

            num_pad = pad_size * (multiplier-1)
            
            # Creates empty arrays half the window size
            # and fills them with either the first or last value
            # If the window size is odd, add 1 extra to the start
            if num_pad % 2 == 0:
                pad_start = np.empty(num_pad // 2)
            else:
                pad_start = np.empty(num_pad // 2 + 1)
            pad_end = np.empty(num_pad // 2)
            pad_start.fill(first_value)
            pad_end.fill(final_value)
            start_series = pd.Series(pad_start)
            end_series = pd.Series(pad_end)

            padded_timeseries = pd.concat([start_series, dfx.iloc[i][j], end_series], axis=0, ignore_index=True) 
            padded_series.append(padded_timeseries)

        padded_series_list.append(padded_series)
       
        padded_frame = pd.DataFrame(padded_series_list)
        padded_frame.columns = dfx.columns
    return padded_frame
```

<br><br>
### Run Rocket Classifier on the augmented dataset



- Combine the orginal training set with the augmented one
- The class labels should remain unchanged when window-warped so the train_y set is appended to itself to match aug_train_x like with the reversed method
- Gives the below accuracy

Accuracy: 85.54621848739495

```python
multiplier = 2
window_size = len(train_x.iloc[0][0])//10

smooth_x = window_warp(train_x, window_size, multiplier)

# Need to pad train_x and test_x to have the same length as the new augmented time series
padded_train_x = pad_series(train_x, window_size, multiplier)
padded_test_x = pad_series(test_x, window_size, multiplier)

smooth_train_x = pd.concat([padded_train_x, smooth_x], axis=0, ignore_index=True)
smooth_test_x = padded_test_x

smooth_train_y = np.append(train_y, train_y, axis=0)

acc, preds = rocket.run_classifier(smooth_train_x, smooth_train_y, padded_test_x, test_y)
print(f"(Accuracy: {acc}")
```

<br><br>
- During implementation, it was found that window warping using the stepped approach in window_warp() produced a better accuracy:

Accuracy: 85.71428571428571

```python
multiplier = 2
window_size = len(train_x.iloc[0][0])//10

stepped_x = window_warp(train_x, window_size, multiplier, True)

# Need to pad train_x and test_x to have the same length as the new augmented time series
padded_train_x = pad_series(train_x, window_size, multiplier)
padded_test_x = pad_series(test_x, window_size, multiplier)

stepped_train_x = pd.concat([padded_train_x, stepped_x], axis=0, ignore_index=True)
stepped_test_x = padded_test_x

stepped_train_y = np.append(train_y, train_y, axis=0)

acc, preds = rocket.run_classifier(stepped_train_x, stepped_train_y, padded_test_x, test_y)
print(f"Accuracy: {acc}")
```

### Compare effect of different mulitpliers on accuracy

- Loop through multiplier values of 3 and 4 (a value of 2 has already been tested above) and run on the ROCKET classifier to test performance
- It is also seen that the stepped approach performs better than the smoothed linear approach

```python
window_size = len(train_x.iloc[0][0])//10

print("Smooth")
for i in range(3,5):
    multiplier = i
    
    warped_x = window_warp(train_x, window_size, multiplier)
    padded_train_x = pad_series(train_x, window_size, multiplier)
    padded_test_x = pad_series(test_x, window_size, multiplier)
    
    aug_train_x = pd.concat([padded_train_x, warped_x], axis=0, ignore_index=True)    
    aug_train_y = np.append(train_y, train_y, axis=0)
    
    acc, preds = rocket.run_classifier(aug_train_x,aug_train_y, padded_test_x, test_y)
    print(f"Multiplier: {multiplier}\nAccuracy: {acc}\n")

print("Stepped")
for i in range(3,5):
    multiplier = i
    
    warped_x = window_warp(train_x, window_size, multiplier, True)
    padded_train_x = pad_series(train_x, window_size, multiplier)
    padded_test_x = pad_series(test_x, window_size, multiplier)
    
    aug_train_x = pd.concat([padded_train_x, warped_x], axis=0, ignore_index=True)    
    aug_train_y = np.append(train_y, train_y, axis=0)
    
    acc, preds = rocket.run_classifier(aug_train_x,aug_train_y, padded_test_x, test_y)
    print(f"Multiplier: {multiplier}\nAccuracy: {acc}\n")
```
